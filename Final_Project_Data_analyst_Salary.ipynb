Add Skills column 

```python
def add_skills_cols(data):
    
    '''Adding skills columns to the dataframe'''
    
    skills_list = ['SQL', 'Excel', 'Tableau', 
              r'\br\b', 'Power BI', 'SAS', 
              'Word', 'PowerPoint', 'AWS', 
              'JavaScript', 'Python']
    
    for skill in skills_list:
        if skill == r'\br\b':
            data['R'] = np.where(data['Job Description'].apply(
                lambda x: x.lower().replace('\n', ' ').replace('  ', ' ')
            ).str.contains(skill), 1, 0)
        else:
            data[skill] = np.where(data['Job Description'].apply(
                lambda x: x.lower().replace('\n', ' ').replace('  ', ' ')
            ).str.contains(skill.lower()), 1, 0)

add_skills_cols(data)
```

```python
data['Size'] = data.loc[~data['Size'].isna(), 'Size'].apply(
    lambda x: x.replace(' employees', '')
)
```

```python
data['Job Title'] = data['Job Title'].replace(
                    ['Sr. Data Analyst', 'Sr Data Analyst', 
                     'Data Analyst Senior', 'Senior Analyst'], 'Senior Data Analyst'
                )

data['Job Title'] = data['Job Title'].replace(
                    ['Jr. Data Analyst', 'Jr Data Analyst', 'Data Analyst Junior'], 'Junior Data Analyst'
            )

data['Job Title'] = data['Job Title'].replace(
                    ['Data Analyst I', 'Data Analyst II', 'Data Analyst III', 
                     'Data Analyst IV', 'Data Analyst V', 'Data Analyst 1',
                     'Data Analyst 2', 'Data Analyst 3', 'Data Analyst 4',
                     'Data Analyst 5', 'Analyst', 'Data Science Analyst'], 'Data Analyst',) 
```

The most in demand skills for data analyst

SQL & Excel are the top 2 most in demand skills for data analysts

```python
skill_count = data[
    ('SQL, Excel, Tableau, R, Power BI, SAS, Word, PowerPoint, AWS, JavaScript, Python').split(', ')
].sum().sort_values(ascending=True)

skill_count_pct = skill_count / len(data) * 100

total_posts = len(data)

fig, ax = plt.subplots(figsize=(8,6))

bars = ax.barh(skill_count_pct.index,
       skill_count_pct.values)

ax.bar_label(bars, fmt='% .2f%%')

ax.set_title('The Most in Demand Skills for Data Analyst Job', y=1.01)
ax.set_xlabel('% of job post')
ax.set_xticks([])

ax.set_xlim(right=75)
```
